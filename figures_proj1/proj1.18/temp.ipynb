```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

plt.rcParams.update({
    "font.family": "serif",
    # "font.serif": ["DejaVu Serif"],
    "mathtext.fontset": "cm",
})
fig, ax = plt.subplots(figsize=(12, 4), dpi = 160)
ax.set(xlim=(0, 14), ylim=(0, 4))
ax.axis("off")

blue, orange = "#5082FF", "#F5820A"
x0, x1 = 1.4, 6.5
yc, yf, yv = 3.5, 2.5, 0.5

ax.plot([x0, x1], [yc, yc], lw=3.0, c=blue)
ax.plot([x0, x1], [yf, yf], lw=3.0, c="black")
ax.plot([x0, x1], [yv, yv], lw=3.0, c=orange)

ax.text(0.45, yc, r"$E_\mathrm{C}$", fontsize=32, va="center")
ax.text(0.45, yf, r"$E_\mathrm{F}$", fontsize=32, va="center")
ax.text(0.45, yv, r"$E_\mathrm{V}$", fontsize=32, va="center")

def double_arrow(x, y1, y2, color):
    ax.add_patch(FancyArrowPatch(
        (x, y1), (x, y2), arrowstyle="<|-|>", mutation_scale=30, lw=3, color=color
    ))

double_arrow(2.3, yv, yf, orange)
ax.text(2.8, (yv + yf) / 2, r"$\Phi_p$", fontsize=32, va="center")

double_arrow(4.55, yf, yc, blue)
ax.text(5.0, (yf + yc) / 2, r"$\Phi_n$", fontsize=32, va="center")

ax.text(7.0, 2.6, r"$\Phi_\mathrm{n} < \Phi_\mathrm{p} $  n-type SB", fontsize=24, va="center")
ax.text(7.0, 1.8, r"$\Phi_\mathrm{n} > \Phi_\mathrm{p} $  p-type SB", fontsize=24, va="center")
ax.text(7.0, 1.0, r"$\Phi_\mathrm{n} $ or $\Phi_\mathrm{p}\sim 0$  Ohmic Barrier", fontsize=24, va="center")

plt.tight_layout()


plt.savefig("proj1.8.pdf")
```
