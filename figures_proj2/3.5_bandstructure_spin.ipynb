# Bandstructure with spin

```python
# Prework

from vmatplot.bandstructure import plot_bandstructure, plot_bandstructure_spin
import matplotlib.pyplot as plt

color_up = "purple"; color_down = "cyan"
dash = (0,(4,3))
# dash = "dashed"
```

```python
spin_list = [["monocolor", "spin up", "3.1_bandstructure/monolayer_FM_HSE06",   color_up, "solid"],
             ["monocolor", "spin down", "3.1_bandstructure/monolayer_FM_HSE06",  color_down, dash],]

plot_bandstructure_spin("Band structure for FM monolayer o-B₁₄", spin_list, "", (-4, 3), True)
plt.savefig("figures/3.23_bs_monolayer_FM.pdf")

fig = plt.gcf()
fig.set_size_inches(9, 6)
plt.savefig("figures_collection/fig2.6a.pdf")
```

```python
spin_list = [["monocolor", "spin up", "3.1_bandstructure/monolayer_AFM_HSE06",   color_up, "solid"],
             ["monocolor", "spin down", "3.1_bandstructure/monolayer_AFM_HSE06", color_down, dash]]

plot_bandstructure_spin("Band structure for AFM monolayer o-B₁₄", spin_list, "", (-4, 3), True)
plt.savefig("figures/3.24_bs_monolayer_AFM.pdf")

fig = plt.gcf()
fig.set_size_inches(9, 6)
plt.savefig("figures_collection/fig2.6b.pdf")
```

```python
bs_list = [["monocolor", "bands", "3.1_bandstructure/bilayer", "blue"]]

plot_bandstructure("Band structure for bilayer o-B₁₄", bs_list, (-4,3), True)
plt.savefig("figures/3.25_bs_bilayer.pdf")

fig = plt.gcf()
fig.set_size_inches(9, 6)
plt.savefig("figures_collection/fig2.6c_PBE.pdf")
```
