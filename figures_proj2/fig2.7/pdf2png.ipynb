# Convert PDF to PNG


```python
import pymupdf
from pathlib import Path

def pdf_to_png(pdf_path, dpi=200):
    """
    Convert each page of a PDF file into PNG images.
    Images are saved in the same folder as the PDF.

    For a single-page PDF:
    example.pdf -> example.png

    For a multi-page PDF:
    example.pdf -> example.png
    example_page_002.png
    example_page_003.png
    """
    pdf_path = Path(pdf_path)

    doc = pymupdf.open(pdf_path)

    zoom = dpi / 72
    matrix = pymupdf.Matrix(zoom, zoom)

    output_files = []

    for page_index in range(len(doc)):
        page = doc[page_index]
        pix = page.get_pixmap(matrix=matrix, alpha=False)

        if page_index == 0:
            output_file = pdf_path.with_suffix(".png")
        else:
            output_file = pdf_path.with_name(
                f"{pdf_path.stem}_page_{page_index + 1:03d}.png"
            )

        pix.save(output_file)
        output_files.append(output_file)

    doc.close()

    print(f"Conversion completed. {len(output_files)} PNG image(s) were created.")

    return output_files
```

```python
pdf_file = "fig2.7_from2.6.pdf"

png_file = pdf_to_png(pdf_file, dpi=1000)
```
