# Plotting program for figure S 2.15


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["lines.solid_capstyle"] = "round"
mpl.rcParams["lines.dash_capstyle"]  = "round"
mpl.rcParams["lines.solid_joinstyle"] = "round"
mpl.rcParams["lines.dash_joinstyle"]  = "round"

figure_size = (10, 6)
figure_dpi = 256

figure_params = {
    "text.usetex": False, "font.family": "serif", "mathtext.fontset": "cm",
    "axes.titlesize": 20, "axes.labelsize": 16, "xtick.labelsize": 14, "ytick.labelsize": 14,
    "legend.fontsize": 12, "figure.facecolor": "w",
    # "pdf.fonttype": 42, "ps.fonttype": 42,
}

plt.rcParams.update(figure_params)

data = [
    ("PBE+D3", 0.0,  0.000),
    ("PBE+D3", 0.5, -0.040),
    ("PBE+D3", 1.0, -0.228),
    ("PBE+D3", 1.5, -0.411),
    ("PBE+D3", 1.8, -0.441),
    ("PBE+D3", 2.0, -0.401),
    ("PBE+D3", 2.5,  0.058),
    ("PBE+D3", 3.0,  0.686),
    ("PBE+D3", 3.5,  1.409),
    ("PBE+D3", 4.0,  2.141),
    (r"r$^2$SCAN", 0.0,  0.000),
    (r"r$^2$SCAN", 0.5, -0.111),
    (r"r$^2$SCAN", 1.0, -0.401),
    (r"r$^2$SCAN", 1.5, -0.685),
    (r"r$^2$SCAN", 1.9, -0.782),
    (r"r$^2$SCAN", 2.0, -0.766),
    (r"r$^2$SCAN", 2.5, -0.311),
    (r"r$^2$SCAN", 3.0,  0.340),
    (r"r$^2$SCAN", 3.5,  1.082),
    (r"r$^2$SCAN", 4.0,  1.834),
]

df = pd.DataFrame(data, columns=["method", "fixed_spin_muB", "relative_energy_eV"])

colors = {"PBE+D3": "#1478E1", r"r$^2$SCAN": "#FA8C00",}
markers = {"PBE+D3": "o", r"r$^2$SCAN": "o",}

fig, ax = plt.subplots(figsize=figure_size, dpi=figure_dpi)

for method in ["PBE+D3", r"r$^2$SCAN"]:
    part = df[df["method"] == method]
    ax.plot(
        part["fixed_spin_muB"],
        part["relative_energy_eV"],
        marker=markers[method],
        color=colors[method],
        linewidth=2.0,
        markersize=8.0,
        label=method,
    )

ax.set_title("Fixed-spin energy profile")
ax.set_xlabel(r"Fixed spin ($\mu_B$)")
ax.set_ylabel("Relative energy (eV)")

ax.set_xlim(-0.05, 4.05)
ax.set_ylim(-0.9, 2.3)
# ax.set_xticks([0, 1, 2, 3, 4])
ax.tick_params(direction="in", which="both", top=True, right=True, bottom=True, left=True)

ax.legend(loc="upper left", frameon=True)

fig.tight_layout()

fig.savefig("S2.15.pdf", bbox_inches="tight")
plt.show()
```
